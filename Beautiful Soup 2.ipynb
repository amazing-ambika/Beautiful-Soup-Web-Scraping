```python
web = requests.get("https://results.eci.gov.in/PcResultGenJune2024/index.htm")

# 
```

```python
soup = BeautifulSoup(web.content,"html.parser")

# html parser for website
```

```python
print(soup.prettify())
```

```python
tables = soup.find_all('table')
# all tables in website
```

```python
tables
# displaying all tables
```

```python
if tables:
    table = tables[0]
    # displaying f
    column_data = table.find_all('tr')
    # making table row into column data
    headers = [header.text.strip() for header in column_data[0].find_all('th')]
    # making the th table heading into headers
    df = pd.DataFrame(columns=headers)
    for row in column_data[1:]:
        # taking column data from 1 and not 0 as taken 0
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        if len(individual_row_data) == len(headers):
            length = len(df)
            # matching header w
            df.loc[length] = individual_row_data
        else:
            print("row does match to headers")
else:
    print("No tables found")
    
    
print(df)
```
